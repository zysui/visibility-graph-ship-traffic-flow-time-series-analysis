```python

import networkx as nx  
from scipy.optimize import curve_fit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
from itertools import combinations  
  
def visibility_graph(series):  
    visibility_graph_edges=[]  
    # convert list of magnitudes into list of tuples that hold the index  
    tseries = []  
    n = 1  
    for magnitude in series:  
        tseries.append( (n, magnitude ) )  
        n += 1  
  
    for a,b in combinations(tseries, 2):  
        # two points, maybe connect  
        (ta, ya) = a  
        (tb, yb) = b  
        connect = True  
        medium=tseries[ta:tb-1]
  
        for tc,yc in medium:  

            if yc > yb + (ya - yb) * ( float(tc - tb) / (ta - tb) ):  
                    connect =False 
        if connect == True:  
            visibility_graph_edges.append((ta,tb))  
    return visibility_graph_edges  
  
def DrawVisibilityGraph(graphEdges,n):  
    plt.figure(figsize=(10,10))
    G=nx.Graph()  #
    for i in graphEdges:  
        G.add_edge(i[0],i[1])  

    pos=nx.spectral_layout(G) #
    nx.draw_networkx_nodes(G,pos,node_size=20)  
    nx.draw_networkx_edges(G,pos,width=1)  
    nx.draw_networkx_labels(G,pos,font_size=2,font_family='sans-serif')  
    #nx.write_gexf(G,'newlabel2.gexf')
    #plt.savefig("dwq.gexf")   #save the picture  
    #plt.show()
    #########################################
    Degree = []
    slist = [x for x in range(1,n+1)] 
    for i in slist:
        d1 = G.degree(i)
        Degree.append(d1)
     
    return Degree
    #plt.savefig("dwq.gexf")   #save the picture  
    #plt.show()
    
####
def nmatrix(n,v_edges):
    shiplist = [x for x in range(1,n+1)] 

    alledges = []
    for i in range(0,len(shiplist)):
        #nex = i+1
        #if i == len(shiplist) - 1:
        #     break
        for j in range(0,len(shiplist)):
            nt = [shiplist[i],shiplist[j]]
            nt = tuple(nt)
            alledges.append(nt)  
    matrix_t = []
    for i in alledges:
        if i in v_edges:
            matrix_t.append(True)

        if i not in v_edges:
            if i[0] == i[1]:
                matrix_t.append(True)
            else:
                matrix_t.append(False)

    #vtx = np.array(matrix_t)#.reshape(len(nn),-1)
    return matrix_t


def alledge(data,c_name):
    bll = []
    for i in c_name:
        c_data = np.array(data[i])
        v_edges = visibility_graph(c_data)

        bll.append(v_edges)
    all_e = []
    for i in bll:
        for j in i:
            if j not in all_e:
                all_e.append(j)
    return all_e

def plot(all_e,d_number):
    #
    figure, ax = plt.subplots()
    labels = ax.get_xticklabels() + ax.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]


    font2 = {'family': 'Times New Roman',
             'weight': 'normal',

             'size': 15,
             }
    
    ax = plt.gca()
    
    plt.xlim([1,d_number])
    plt.ylim([d_number,1])
    
    for i in all_e:
        plt.plot(i[0],i[1],'ks',markersize = 2)
        plt.plot(i[1],i[0],'ks',markersize = 2)
    for j in range(1,d_number+1):
        plt.plot(j,j,'ks',markersize = 2)

 
    plt.xlabel('Node label',font2)
    plt.ylabel('Node label',font2)
    plt.grid()
    return plt.savefig("E:/python_code/pythonworkspace/situation_network/可视图分析—20191023/newlabel2.png") 
```
